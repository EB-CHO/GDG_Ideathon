```python
import pandas as pd

df = pd.read_csv('./train.csv')
df
```

```python
df['남은인원'] = df['본사정원수'] - df['본사휴가자수'] - df['본사출장자수'] - df['현본사소속재택근무자수']
df['석식비율'] = df['석식계'] / df['남은인원']
```

```python
df = df.drop(columns=['본사휴가자수', '본사출장자수', '본사시간외근무명령서승인건수', '현본사소속재택근무자수', '조식메뉴', '중식메뉴', '중식계'])
df
```

```python
df.columns = ['date','day','total','dinner','people', 'left', 'percentage']
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df['yearmonth'] = df['date'].dt.strftime('%Y%m')
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%m%d')
```

```python
df['Month'] = df['date'].str[0:2]
df['Date'] = df['date'].str[2:]
df.drop(['date'], inplace=True , axis=1) # 아침은 영향이 없을것같으므로 삭제

df.head(2)
```

```python
dinner_menu = []
for day in range(len(df)):
    tmp = df.loc[day,'dinner'].split(' ')
    tmp = ' '.join(tmp).split()

    for menu in tmp:
        if '(' in menu:
            tmp.remove(menu)
    dinner_menu.append(tmp)
```

```python
df['dinner_menu'] = dinner_menu
```

```python
len(df['dinner_menu'])
```

```python
dinner_menu[0:5]
```

```python
print(range(len(df)))
print(range(len(dinner_menu)))
```

```python
bobd=[]
soupd=[]
maind=[]


for word in  dinner_menu:


    if len(word) == 0:
        bobd.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        soupd.append('None')
        maind.append('None')
    elif '*' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')
    elif '가정의날' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')
    elif '가정의달' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')
    elif '자기계발의날' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')
    elif '*자기계발의날*' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')
    elif '자기개발의날' in word:
        bobd.append('None')
        soupd.append('None')
        maind.append('None')

    else:
        bobd.append(word[0])
        soupd.append(word[1])
        maind.append(word[2])
```

```python
#다들어왔는지 확인
print(len(dinner_menu))
print(len(bobd))
print(len(soupd))
print(len(maind))
```

```python
df['bobd'] = bobd
df['soupd'] = soupd
df['maind'] = maind

df['maind'].value_counts()[:10]
```

```python
train = df[
    ['day','total','people','left', 'percentage','Month','Date','bobd','soupd','maind']
]
```

```python
train = train[~train['bobd'].str.contains("None", na=False, case=False)]
```

```python
train.head()
```

```python
# encoding

train['day'] =  train['day'].astype('category')
train['day'] = train.day.cat.codes


train['bobd'] =  train['bobd'].astype('category')
train['bobd'] = train.bobd.cat.codes

train['soupd'] =  train['soupd'].astype('category')
train['soupd'] = train.soupd.cat.codes

train['maind'] =  train['maind'].astype('category')
train['maind'] = train.maind.cat.codes
```

```python
train
```

```python
train.to_csv('./data.csv')
```

```python
train['Month'] = train['Month'].astype(int)
train['Date'] = train['Date'].astype(int)
```

```python
features = ['bobd', 'soupd', 'maind', 'day', 'Month', 'Date']
target_people = 'people'
target_percentage = 'percentage'
```

```python
# 특성과 타겟 분리
X = train[features]
y_people = train[target_people]
y_percentage = train[target_percentage]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

# 데이터셋 분할 (학습 80%, 테스트 20%)
X_train, X_test, y_people_train, y_people_test = train_test_split(X, y_people, test_size=0.2, random_state=42)
X_train, X_test, y_percentage_train, y_percentage_test = train_test_split(X, y_percentage, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor(random_state=42)
catboost_model = CatBoostRegressor(random_state=42, verbose=0)
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42, enable_categorical=True)
```

```python
# 모델 학습 및 예측 - RandomForest
rf_model.fit(X_train, y_people_train)
y_people_pred_rf = rf_model.predict(X_test)
rf_mse_people = mean_squared_error(y_people_test, y_people_pred_rf)

rf_model.fit(X_train, y_percentage_train)
y_percentage_pred_rf = rf_model.predict(X_test)
rf_mse_percentage = mean_squared_error(y_percentage_test, y_percentage_pred_rf)

print(f'RandomForest MSE (people): {rf_mse_people}')
print(f'RandomForest MSE (percentage): {rf_mse_percentage}')
```

```python
# 모델 학습 및 예측 - CatBoost
catboost_model.fit(X_train, y_people_train)
y_people_pred_catboost = catboost_model.predict(X_test)
catboost_mse_people = mean_squared_error(y_people_test, y_people_pred_catboost)

catboost_model.fit(X_train, y_percentage_train)
y_percentage_pred_catboost = catboost_model.predict(X_test)
catboost_mse_percentage = mean_squared_error(y_percentage_test, y_percentage_pred_catboost)

print(f'CatBoost MSE (people): {catboost_mse_people}')
print(f'CatBoost MSE (percentage): {catboost_mse_percentage}')
```

```python
# GridSearch를 이용한 XGBoost 하이퍼파라미터 튜닝
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')

# 사람 수 예측을 위한 GridSearchCV 학습
grid_search.fit(X_train, y_people_train)
y_people_pred_xgb = grid_search.best_estimator_.predict(X_test)
xgb_mse_people = mean_squared_error(y_people_test, y_people_pred_xgb)

# 비율 예측을 위한 GridSearchCV 학습
grid_search.fit(X_train, y_percentage_train)
y_percentage_pred_xgb = grid_search.best_estimator_.predict(X_test)
xgb_mse_percentage = mean_squared_error(y_percentage_test, y_percentage_pred_xgb)

print(f'XGBoost MSE (people): {xgb_mse_people}')
print(f'XGBoost MSE (percentage): {xgb_mse_percentage}')
```
